```python
dscu = dsc.merge(dsu, on=None, how='left', left_index=True, right_index=True)
```

```python
dscug = dscu.groupby(dscu.index)
```

```python
TRAJET = [
'domicile/travail',
'domicile/école',
'courses/achats',
'professionnel',
'de loisirs'
]
```

```python
def _gen_sentence_u(a,s,g,c,u,d):
    grav = ('indemne', 'tué(e)', 'blessé(e) hospitalisé(e)', 'blessé(e) léger')[g-1]
    try:
        age = int(a)
    except:
        age = 0
    if age<13:
        w = 'garçon' if s==1 else 'fille'
    elif age<18:
        w = 'adolescent' if s==1 else 'adolescente'
    else:
        w = 'homme' if s==1 else 'femme'
    if s==1:
        tp = ('conducteur', 'passager', 'piéton', 'à roller ou trottinette')[c-1]
    else:
        tp = ('conductrice', 'passagère', 'piétonne', 'à roller ou trottinette')[c-1]
    if u[0] == '1':
        if u[1] == '1':
            ceinture = ' ceinturé(e)'
        elif u[1] == '2':
            ceinture = ' non ceinturé(e)'
        else:
            ceinture = ''
    else:
        ceinture = ''
    s = "{0:s}, {2:d} ans, {1:s}, {3:s}{4:s}".format(w, grav, age, tp, ceinture)
    if 1 <= d <= 5:
        s += ', déplacement ' + TRAJET[int(d)-1]
    return s
```

```python
def f(x):
    age = 2012-x['an_nais'].values
    sexe = x['sexe'].values
    grav = x['grav'].values
    catu = x['catu'].values
    trajet = x['trajet'].values
    secu = x['secu'].astype(str).values
    return '<br/>'.join(_gen_sentence_u(*d) for d in zip(age, sexe, grav, catu, secu, trajet))
dscugagg = dscug.apply(f)
```

```python
ds['details_usagers'] = dsc['details_usagers'] = dscugagg
```

```python
MANV = [
'Sans changement de direction',
'Même sens, même file',
'Entre 2 files',
'En marche arrière',
'A contresens',
'En franchissant le terre-plein central',
'Dans le couloir bus, dans le même sens',
'Dans le couloir bus, dans le sens inverse',
'En s\'insérant',
'En faisant demi-tour sur la chaussée',
'Changeant de file à gauche',
'Changeant de file à droite',
'Déporté à gauche',
'Déporté à droite',
'Tournant à gauche',
'Tournant à droite',
'Dépassant à gauche',
'Dépassant à droite',
'Traversant la chaussée',
'Manoeuvre de stationnement',
'Manoeuvre d\'évitement',
'Ouverture de porte',
'Arrêté (hors stationnement)',
'En stationnement (avec occupants)',
]
```

```python
def _gen_sentence_v(catv,manv):
    try:
        m = ', ' + MANV[int(manv)-1].lower()
    except ValueError:
        m = ''
    return CAT.get(catv, 'autre') + m
```

```python
def g(x):
    catv = x['catv'].values
    manv = x['manv'].values
    return '<br/>'.join(_gen_sentence_v(c,m) for c,m in zip(catv, manv))
dsc['details_vehicules'] = dscvg.apply(g)
dsc['details_vehicules'].tail(3)
```

## champs indices ==> textuels

```python
def _atm(x):
    try:
        x = int(x)
    except ValueError:
        x = 0
    return ATM[(x)-1] if 1<=(x) else ''
ATM = ['Normale', 'Pluie légère', 'Pluie forte', 'Neige - grêle', 
       'Brouillard', 'Vent fort - tempête', 'Temps éblouissant', 'Temps couvert',
       'Autre']
ds['atm_text'] = ds['atm'].apply(_atm)
ds['atm_text'].tail()
```

```python
LUM = ['Plein jour', 'Crépuscule ou aube', 'Nuit sans éclairage public',
       'Nuit avec éclairage public non allumé', 'Nuit avec éclairage public allumé']
ds['lum_text'] = ds['lum'].apply(lambda x: LUM[int(x)-1] if 1<=int(x) else '')
ds['lum_text'].tail()
```

```python
CATR = ['Autoroute', 'Route nationale', 'Route départementale',
        'Voie communale', 'Hors réseau public', 
        'Parc de stationnement ouvert à la circulation publique',]
ds['catr_text'] = ds['catr'].apply(lambda x: CATR[int(x)-1] if 1<=int(x)<=6 else '')
ds['catr_text'].tail()
```

```python
INT = ['Hors intersection',
        'Intersection en X',
        'Intersection en T',
        'Intersection en Y',
        'Intersection à plus de 4 branches',
        'Giratoire',
        'Place',
        'Passage à niveau',
        'Autre intersection',]
ds['int_text'] = ds['int'].apply(lambda x: INT[int(x)-1] if 1<=int(x) else '')
ds['int_text'].tail()
```

```python
INFRA = [
'Souterrain - tunnel',
'Pont - autopont',
'Bretelle d\'échangeur ou de raccordement',
'Voie ferrée',
'Carrefour aménagé',
'Zone piétonne',
'Zone de péage',
]
def _infra(x):
    try:
        x = int(x)
    except ValueError:
        x = 0
    return INFRA[(x)-1] if 1<=(x) else ''
ds['infra_text'] = ds['infra'].apply(_infra)
ds['infra_text'].tail()
```

```python
COLLISION = [
'collision frontale',
'collision par l\'arrière',
'collision par le côté',
'collision en chaîne',
'collisions multiples'
]
ds['col_text'] = ds['col'].apply(lambda x: COLLISION[int(x)-1] 
                                             if 1<=int(x)<=len(COLLISION) else '')
ds['col_text'].tail()
```
