## Load databases

```python
dsc = pd.read_csv('data2012/caracteristiques.csv', index_col=0)
dsc.head(2)
```

```python
N = len(dsc)
```

```python
dsv = pd.read_csv('data2012/vehicules.csv', index_col=0)
dsv.head(2)
```

```python
dsu = pd.read_csv('data2012/usagers.csv', index_col=0)
dsu.head(2)
```

```python
dsl = pd.read_csv('data2012/lieux.csv', index_col=0)
dsl.head(2)
```

## Clean Geoloc

```python
dsc.lat[dsc.lat.isnull()] = 0
```

```python
dsc.long[dsc.long.isnull()] = 0
```

```python
dsc['lat'] = dsc.lat.astype(np.float32)/1e5
dsc['long'] = dsc.long.str.replace('E','').astype(np.float32)/1e5
```

## Injured

```python
dsc.head(2)
```

```python
dscu = dsc.merge(dsu, on=None, how='left', left_index=True, right_index=True)
```

```python
dscu.head(2)
```

```python
dscug = dscu.groupby(dscu.index)
```

```python
def _agg_gravite(x):
    if 2 in x.values:
        return 2
    elif 3 in x.values:
        return 1
    else:
        return 0
```

```python
dsc['gravite'] = dscug.agg({'grav': _agg_gravite})
```

```python
dsc.tail(2)
```

## Nbimpliques

```python
dscv = dsc.merge(dsv, on=None, how='left', left_index=True, right_index=True)
```

```python
dscvg = dscv.groupby(dscv.index)
```

```python
dsc['nbimpliques'] = dscvg.agg({'an': lambda x: x.count()})
```

```python
dsc.head(4)
```

```python
dscu.columns
```

## Catégorie age

```python
years = [1932, 1952, 1972, 1992, 2020]
def _agg_an_nais(i, x):
    return np.bincount( np.digitize(x.values, years), minlength=5)[i]
```

```python
for i in range(5):
    dsc['age_'+str(i)] = dscug.agg({'an_nais': functools.partial(_agg_an_nais, i)})
```

```python
dsc.head(2)
```

## Sexe

```python
dsc['hommes'] = dscug.agg({'sexe': lambda x: np.sum(x.values==1)})
dsc['femmes'] = dscug.agg({'sexe': lambda x: np.sum(x.values==2)})
```

```python
dsc.head(3)
```

## Merge avec lieux

```python
dscl = dsc.merge(dsl, on=None, how='left', left_index=True, right_index=True)
```

```python
dsclg = dscl.groupby(dscl.index)
```

```python
dsclgmax = dsclg.max()
```

```python
for _f in (
        'catr',
        'circ',
        'nbv',
        'vosp',
        'prof',
        'plan',
        'lartpc',
        'larrout',
        'surf',
        'infra',
        'situ',
        'env1',
        ):
    dsc[_f] = dsclgmax[_f]
```

```python
dsc['env1'] = dsc['env1']>0
```

```python
dsc.head(3)
```

## Code Insee

```python
insee.head(2)
```

```python
dsc.tail(2)
```

## Jour de la semaine

```python
year = '20' + dsc.an.astype(str)
month = dsc.mois.astype(str).str.pad(2, side='left').str.replace(' ', '0')
day = dsc.jour.astype(str).str.pad(2, side='left').str.replace(' ', '0')
hrmn = dsc.hrmn.astype(str).str.pad(4, side='left').str.replace(' ', '0')
hh = hrmn.str.slice(0,2)
mm = hrmn.str.slice(2,4)
```

```python
date = pd.to_datetime(year+month+day+' '+hh+':'+mm)
```

```python
dsc['date'] = date
```

```python
dsc['dayofweek'] = pd.DatetimeIndex(date).dayofweek
```
