```python
import functools, operator, itertools
import locale
import sys
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import prettyplotlib as ppl
%matplotlib inline
mpl.rcParams['savefig.dpi'] = 100
```

## Chargement des bases de données

```python
dsc = pd.read_csv('data2012/caracteristiques.csv', index_col=0)
```

```python
N = len(dsc)
```

```python
dsv = pd.read_csv('data2012/vehicules.csv', index_col=0)
```

```python
dsu = pd.read_csv('data2012/usagers.csv', index_col=0)
```

```python
dsl = pd.read_csv('data2012/lieux.csv', index_col=0)
```

On ajoute quelques champs utiles.

L'âge

```python
dsu['age'] = 2012-dsu['an_nais']
```

et le jour de la semaine

```python
year = '20' + dsc.an.astype(str)
month = dsc.mois.astype(str).str.pad(2, side='left').str.replace(' ', '0')
day = dsc.jour.astype(str).str.pad(2, side='left').str.replace(' ', '0')
hrmn = dsc.hrmn.astype(str).str.pad(4, side='left').str.replace(' ', '0')
hh = hrmn.str.slice(0,2)
mm = hrmn.str.slice(2,4)
date = pd.to_datetime(year+month+day+' '+hh+':'+mm)
dsc['dayofweek'] = pd.DatetimeIndex(date).dayofweek
```

## Sélection des jeunes tués sur la route en 2012

```python
agemin, agemax = 15, 25
```

```python
dsuc = dsu.merge(dsc, left_index=True, right_index=True, how='left') \
          .merge(dsl, left_index=True, right_index=True, how='left')
```

```python
jt = ((agemin <= dsuc['age']) & (dsuc['age'] < agemax) & 
      (dsuc['grav'] == 2) &
      (dsuc['catu'] <= 2)  # on enlève les piétons
      )
```

```python
djt = dsuc[jt]
```

```python
print("Il y a eu {n} jeunes tués sur la route en 2012.".format(n=jt.sum()))
```

Quels étaient leurs âges?

```python
ppl.hist(djt.age.values, bins=range(agemin, agemax+1));
plt.xlim(agemin, agemax);
plt.xlabel('Âge');
plt.ylabel('Nombre de victimes');
```

Moment de l'accident?

```python
ax = ppl.hist(djt.dayofweek.values, bins=range(0, 8));
ax.set_xticklabels(['Lundi', 'Mardi', 'Mercredi', 'Jeudi', 
                    'Vendredi', 'Samedi', 'Dimanche']);
ax.set_xticks(np.arange(7)+.5);
plt.ylabel('Nombre de victimes');
```

```python
hour = (djt.hrmn)//100
```

```python
ax = ppl.hist(hour.values, bins=np.arange(25));
plt.xlabel('Heure');
plt.ylabel('Nombre de victimes');
plt.xlim(0, 23);
```

```python
hour_week = djt.dayofweek + hour/24.
```

```python
ax = ppl.hist(hour_week.values, bins=np.arange(0., 7.01, 1./12),
              lw=0);
plt.ylabel('Nombre de victimes');
plt.vlines(np.arange(7), 0, 30, alpha=.5);
ax.set_xticklabels(['Lundi', 'Mardi', 'Mercredi', 'Jeudi', 
                    'Vendredi', 'Samedi', 'Dimanche']);
ax.set_xticks(np.arange(7)+.5);
plt.ylabel('Nombre de victimes');
plt.ylim(0, 30);
```

## Circonstances de ces accidents

```python
djt.columns
```

Feature selection:
* lum: 1 (jour) ou >=2
* agg
* atm: [1,2] (normal) ou >=3
* catr [<=4]
* catu: 1 (conducteur) 2 (passager) 3 (piéton)
* (manv)

```python
fet = djt[[
           'lum', 
           'agg', 
           #'atm', 
           'catr', 
           'catu'
           ]]
fet['lum'] = (fet['lum']==1).astype(np.int8)
#fet['atm'] = (fet['atm']<=2).astype(np.int8)
fet['catr'][fet['catr']>=5] = 0
fet.describe()
```

```python
fet.hist();
```

```python
plt.figure(figsize=(8,8));
a = fet.values + np.random.random_sample(fet.shape)*.4
for i, j in itertools.product(range(4), range(4)):
    if j!=i:
        plt.subplot(4,4,1+4*i+j)
        ppl.scatter(a[:,j], a[:,i], c=fet.lum, alpha=.1);
        plt.xlabel(fet.columns[j]);
        plt.ylabel(fet.columns[i]);
plt.tight_layout()
```
