### On reprend le fichier principal

```python
ds = pd.read_csv('data2012/final2.csv', index_col=0)
```

On vire les champs foireux.

```python
del ds['postal_code']
del ds['nom_comm']
del ds['nom_dept']
del ds['nom_region']
```

```python
ds.head(2)
```

On choisit les points dont on veut mettre à jour les coordonnées.

```python
# on maj BANO et centroid only
maj = (ds_brice.geocodtype == 'BANO') | (ds_brice.geocodtype == 'centroid')
# et que lorsqu'on a pas déjà des coordonnées
maj = maj & ~check(ds.lat, ds.long)
maj.sum()
```

On ajoute qq champs.

```python
ds['geocodtype'] = ds_brice.geocodtype
```

```python
ds['insee_com'] = ds_brice.insee_com.astype(str)
```

On met à jours les coordonnées.

```python
ds['lat'][maj] = ds_brice[maj].nlat  # attention, ['lat'][maj] l'ordre est important
ds['long'][maj] = ds_brice[maj].nlng
```

On ajoute localisation approximative pour les centroïdes.

```python
ds['approx'] = np.zeros(len(ds), dtype=str)
ds['approx'][ds.geocodtype == 'centroid'] = 'Localisation approximative'
```

```python
ds.head(2)
```

### Rechargement des codes INSEE

On charge aussi le fichier des codes INSEE.

```python
insee = pd.read_csv('data/code-insee-code-postal.csv', delimiter=';')
```

```python
insee.head(2)
```

```python
villes = pd.merge(ds.reset_index(), insee, on='insee_com',  how='left').set_index('Num_Acc')
```

```python
for f in ('postal_code', 'nom_comm', 'nom_dept', 'nom_region'):
    ds[f] = villes[f]
```

```python
ds.head(3)
```

Enfin, on vérifie combien de coordonnées on a.

```python
avirer = ~check(ds.lat, ds.long)
ds.lat[avirer] = np.nan
ds.long[avirer] = np.nan
```
