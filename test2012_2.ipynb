## FIXING Code Insee

```python
insee = pd.read_csv('data/code-insee-code-postal.csv', delimiter=';')
```

```python
insee.head(2)
```

```python
dsc['insee_com'] = dsc.dep.astype(str).str.slice(0,2).str.pad(2, side='right').str.replace(' ', '0') + \
    dsc.com.astype(str).str.slice(0, 3).str.pad(3, side='right').str.replace(' ', '0')
```

```python
dsc['insee_com'].head()
```

```python
for f in ('postal_code','nom_comm','nom_dept','nom_region'):
    del dsc[f]
```

```python
villes = pd.merge( dsc.reset_index(), insee, on='insee_com',  how='left').set_index('Num_Acc')
villes.tail(3)
```

```python
for f in ('postal_code','nom_comm','nom_dept','nom_region'):
    dsc[f] = villes[f]
```
